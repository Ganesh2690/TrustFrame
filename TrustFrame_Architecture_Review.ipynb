## Section 2 — Repository Ingestion and Tree Traversal

````python
GENERATED_DIRS = {"__pycache__", "node_modules", ".git", "dist", ".pytest_cache", ".mypy_cache", "build", ".venv", "venv", "env"}
GENERATED_EXTS = {".pyc", ".pyo"}

def walk_tree(root: pathlib.Path, prefix: str = "", depth: int = 0, max_depth: int = 6):
    """Print an annotated tree, skipping generated folders."""
    entries = sorted(root.iterdir(), key=lambda p: (p.is_file(), p.name.lower()))
    for i, entry in enumerate(entries):
        is_last = i == len(entries) - 1
        connector = "└── " if is_last else "├── "
        if entry.name in GENERATED_DIRS:
            print(f"{prefix}{connector}{entry.name}/  [GENERATED — skipped]")
            continue
        if entry.suffix in GENERATED_EXTS:
            continue
        tag = ""
        if entry.is_dir():
            print(f"{prefix}{connector}{entry.name}/")
            extension = "    " if is_last else "│   "
            if depth < max_depth:
                walk_tree(entry, prefix + extension, depth + 1, max_depth)
        else:
            # Annotate key files
            notes = {
                "main.py": "← API entry point",
                "config.py": "← Settings loader",
                "database.py": "← SQLAlchemy engine",
                "requirements.txt": "← Python deps",
                "package.json": "← Node deps",
                "docker-compose.yml": "← Compose orchestration",
                "Dockerfile": "← Container image",
                "vite.config.js": "← Vite + dev proxy",
                "alembic.ini": "← Alembic config",
                ".env": "← Local secrets (DO NOT COMMIT)",
                ".env.example": "← Env template",
                "nginx.conf": "← Production reverse proxy",
            }
            tag = f"  {notes[entry.name]}" if entry.name in notes else ""
            print(f"{prefix}{connector}{entry.name}{tag}")

print("=" * 70)
print("TrustFrame REPOSITORY TREE")
print("=" * 70)
walk_tree(REPO_ROOT)
````
